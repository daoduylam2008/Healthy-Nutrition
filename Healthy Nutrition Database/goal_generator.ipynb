```python
import pandas
import json
```

# US Style Diet

```python
us_style_diet = pandas.read_excel("assets/Diet Chart.xlsx", sheet_name="U.S.-Style Diet Chart")
us_style_diet
```

```python
us_style_diet_jsons = []

for i in range(1, 17):
    data = {
        "name": "US Style Diet",
        "calorie": us_style_diet["Calorie"][i] * 1000.0,
        "carbs": us_style_diet["Carbs.1"][i],
        "protein": us_style_diet["Protein.1"][i],
        "fat": us_style_diet["Fat.1"][i]
    }
    us_style_diet_jsons.append(data)
```
