```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imutils
```

```python
img = cv2.imread("image.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
(h,w) = img.shape[:2]
```

```python
plt.imshow(img)
plt.show()
```

# Rotate Image

```python
matrix = cv2.getRotationMatrix2D((w / 2, h / 2), 90, 1)
rotate_90_img = cv2.warpAffine(img, matrix, (w, h))
```

```python
plt.imshow(rotate_90_img)
plt.show()
```

# Resize image

```python
scale = 0.5
resize_img = cv2.resize(img, (int(w * scale), int(h * scale)))

plt.imshow(resize_img)
plt.show()
```

# Flip image

```python
flipped_image_1 = cv2.flip(img, 1)
plt.imshow(flipped_image_1)
plt.show()
```

```python
flipped_image_0 = cv2.flip(img, 0)
plt.imshow(flipped_image_0)
plt.show()
```

```python
flipped_image_n1 = cv2.flip(img, -1)
plt.imshow(flipped_image_n1)
plt.show()
```

# Translate image

```python
shift_x = -50
shift_y = -90

M = np.float32(
    [[
        1, 0, shift_x
    ],
    [
        0, 1, shift_y
    ]]
)
```

```python
translated_image = cv2.warpAffine(img, M, (w, h))
plt.imshow(translated_image)
plt.show()
```

```python
translated_image = imutils.translate(img, -50, -90)
plt.imshow(translated_image)
plt.show()
```
